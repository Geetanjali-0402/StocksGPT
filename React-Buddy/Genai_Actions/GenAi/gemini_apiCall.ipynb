```python
import os

def find_file_upwards(starting_path, target_filename):
    # Set the current path to the starting path
    current_path = starting_path

    while True:
        # Check if the target file exists in the current directory
        for root, dirs, files in os.walk(current_path):
            if target_filename in files:
                return os.path.join(root, target_filename)
        
        # Move up one directory level
        parent_path = os.path.dirname(current_path)
        
        # If we have reached the root directory, stop searching
        if parent_path == current_path:
            break
        
        current_path = parent_path

    # Return None if the file was not found
    return None

# Example usage
starting_path = "C:\\Users\\DGASERLAP177\\Documents\\GitHub\\StocksGPT\\src\\pages\\Authentication" # Starting from the current directory
target_filename = "actions.ts"  # The name of the file to search for

found_file_path = find_file_upwards(starting_path, target_filename)

if found_file_path:
    print(f"File found: {found_file_path}")
    # You can also perform additional actions with the found file path
else:
    print("File not found")
```

```python
import os
import json

def fileNavigator(FolderPathPython, JsRelativePath):
    # Extracting the number of upward folders and the relative folder path from JsRelativePath
    split_path = JsRelativePath.split('/')
    NoOfUpFolders = sum(1 for part in split_path if part == '..')
    JsRelativeFolderPath = '/'.join(part for part in split_path if part not in ['..', '.'])
    
    # Converting JsRelativeFolderPath to Python relative folder path
    PyRelativeFolderPath = JsRelativeFolderPath.replace('/', '\\')
    
    # Splitting FolderPathPython into components
    folder_components = FolderPathPython.split('\\')
    
    # Trimming the folder path based on NoOfUpFolders
    if NoOfUpFolders > 0:
        TrimmedFolderPathPython = '\\'.join(folder_components[:-NoOfUpFolders])
    else:
        TrimmedFolderPathPython = FolderPathPython
    
    # Constructing the new file path
    NewFilePath = os.path.join(TrimmedFolderPathPython, PyRelativeFolderPath)
    
    return NewFilePath

# Function to read JS relative paths from a JSON file and process each
def processPathsFromJson(jsonFilePath, FolderPathPython):
    # Read the JSON file
    with open(jsonFilePath, 'r') as file:
        jsPaths = json.load(file)
    
    # Initialize an empty dictionary to store the new file paths
    file_paths_dict = {}

    # Process each component's JS relative path and store the new file path in the dictionary
    for component in jsPaths["components"]:
        name = component["name"]
        JsRelativePath = component["path"].strip('`')
        new_file_path = fileNavigator(FolderPathPython, JsRelativePath)
        new_file_path_with_extension = new_file_path + '.tsx'
        # Store in dictionary
        file_paths_dict[name] = new_file_path_with_extension

    return file_paths_dict

# Function to print the contents of each file path
def printFileContents(file_paths_dict):
    for component_name, file_path in file_paths_dict.items():
        if os.path.exists(file_path):
            with open(file_path, 'r') as file:
                content = file.read()
                print(f"Contents of {component_name} ({file_path}):\n{content}\n")
        else:
            print(f"File {file_path} does not exist.\n")

# Example usage
FolderPathPython = "C:\\Users\\DGASERLAP177\\Documents\\GitHub\\StocksGPT\\src\\pages\\Authentication"
jsonFilePath = "components_login.json"  # Assuming this file contains the JSON array of paths

# Process paths and store the result in a dictionary
result = processPathsFromJson(jsonFilePath, FolderPathPython)

# Print the result dictionary
print("Dictionary of new file paths:")
for key, value in result.items():
    print(f"{key}: {value}")

# Print the contents of each file path
printFileContents(result)
```
